```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Figure 7**

```python
data = pd.read_csv("kappa_u_0_3_Chern_L_50_APBC.csv", header = None)
u_APBC = data[0][::3]
kappa_APBC = data[1][::3]
G_APBC = data[2][::3]
R2_APBC = data[3][::3]
```

```python
data = pd.read_csv("kappa_u_0_3_Chern_L_50_PBC.csv", header = None)
u_PBC = data[0][::3]
kappa_PBC = data[1][::3]
G_PBC = data[2][::3]
R2_PBC = data[3][::3]
```

Fig 7(b)

```python
plt.plot(u_APBC, G_APBC, ".", label = "APBC")
plt.plot(u_PBC, G_PBC, ".", label = "PBC")
plt.xlabel("u", fontsize=20)
plt.ylabel(r"$G(A:B:C)$", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=16)
plt.tick_params(axis='both', direction='in')
plt.ylim(-1, 57)
plt.xticks([0, 0.5, 1, 1.5, 2, 2.5, 3], labels=["0.", "0.5", "1.", "1.5", "2.", "2.5", "3."])
plt.yticks([0, 10, 20, 30, 40, 50], labels=["0", "10", "20", "30", "40", "50"])
plt.show()
```

Fig 7(c)

```python
plt.plot(u_APBC, kappa_APBC*8/np.log(2), ".", label = "APBC")
plt.plot(u_PBC, kappa_PBC*8/np.log(2), ".", label = "PBC")
plt.xlabel("u", fontsize=20)
plt.ylabel(r"$8\kappa/\ln2$", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=16)
plt.xticks([0, 0.5, 1, 1.5, 2, 2.5, 3], labels=["0.", "0.5", "1.", "1.5", "2.", "2.5", "3."])
plt.tick_params(axis='both', direction='in')
plt.show()
```
